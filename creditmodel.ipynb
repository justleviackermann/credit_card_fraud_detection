```python
from scipy.stats import norm    
f,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(25,6))
v14fraud=new_df['V14'].loc[new_df['Class']==1].values
sns.histplot(v14fraud,ax=ax1,kde=True,color='red',alpha=0.6,stat="density")
ax1.set_title("V14 Distribution \n (Fraud Transactions")
mean,sigma=norm.fit(v14fraud)
x = np.linspace(v14fraud.min(), v14fraud.max(), 200) # create smooth x-axis
y = norm.pdf(x, mean, sigma) 
ax1.plot(x, y, 'k', lw=2)

ax1.set_title("V12 Distribution \n (Fraud Transactions")
v12fraud=new_df['V12'].loc[new_df['Class']==1].values
sns.histplot(v12fraud,ax=ax2,kde=True,color='blue',alpha=0.6,stat="density")
ax2.set_title("V12 Distribution \n (Fraud Transactions")
mean,sigma=norm.fit(v12fraud)
x = np.linspace(v12fraud.min()-2, v12fraud.max()+2, 200) # create smooth x-axis
y = norm.pdf(x, mean, sigma) 
ax2.plot(x, y, 'k', lw=2)

ax1.set_title("V10 Distribution \n (Fraud Transactions")
v10fraud=new_df['V10'].loc[new_df['Class']==1].values
sns.histplot(v10fraud,ax=ax3,kde=True,color='green' ,alpha=0.6,stat="density")
ax3.set_title("V10 Distribution \n (Fraud Transactions")
mean,sigma=norm.fit(v10fraud)
x = np.linspace(v10fraud.min(), v10fraud.max(), 200) # create smooth x-axis
y = norm.pdf(x, mean, sigma) 
ax3.plot(x, y, 'k', lw=2)
```

we can see that the curve is skewed to left . so these are easy for the ml models to interpret and we cant use naive bayes as it expects the curve to be perfect bell..
models like logistic regression,decision tree,svm might perform well   

```python
#Removing outliers from V14
v14_fraud = new_df['V14'].loc[new_df['Class'] == 1].values
q25, q75 = np.percentile(v14_fraud, 25), np.percentile(v14_fraud, 75)
print(' 25% of v14_fraud : {} | 75% of v14_fraud: {}'.format(q25, q75))
v14_iqr = q75 - q25
print('iqr: {}'.format(v14_iqr))

v14_cutoff = v14_iqr * 1.5
v14_lower, v14_upper = q25 - v14_cutoff, q75 + v14_cutoff
print('Cut Off: {}'.format(round(v14_cutoff,2)))
print('V14 Lower: {}'.format(round(v14_lower,2)))
print('V14 Upper: {}'.format(round(v14_upper,2)))
#condition to remove an outlier
outliers = [x for x in v14_fraud if x < v14_lower or x > v14_upper]
print('Number of outliers in V14 for Fraud Cases: {}'.format(len(outliers)))
print('V14 outliers:', [float(x) for x in outliers])
new_df = new_df.drop(new_df[(new_df['V14'] > v14_upper) | (new_df['V14'] < v14_lower)].index)
#Number of instances after removing the outliers
print('Number of Instances after outliers removal: {}'.format(len(new_df)))
print('----' * 20)


#Removing outliers from V12
v12_fraud = new_df['V12'].loc[new_df['Class'] == 1].values
q25, q75 = np.percentile(v12_fraud, 25), np.percentile(v12_fraud, 75)
print(' 25% of v12_fraud : {} | 75% of v12_fraud: {}'.format(q25, q75))
v12_iqr = q75 - q25
print('iqr: {}'.format(v12_iqr))
v12_cutoff = v12_iqr * 1.5
v12_lower, v12_upper = q25 - v12_cutoff, q75 + v12_cutoff
print('Cut Off: {}'.format(round(v12_cutoff,2)))
print('V12 Lower: {}'.format(round(v12_lower,2)))
print('V12 Upper: {}'.format(round(v12_upper,2)))
#condition to remove an outlier
outliers = [x for x in v12_fraud if x < v12_lower or x > v12_upper]
print('Number of outliers in V12 for Fraud Cases: {}'.format(len(outliers)))
print('V12 outliers:', [float(x) for x in outliers])
new_df = new_df.drop(new_df[(new_df['V12'] > v12_upper) | (new_df['V12'] < v12_lower)].index)
#Number of instances after removing the outliers
print('Number of Instances after outliers removal: {}'.format(len(new_df)))
print('----' * 20)



#Removing outliers from V10
v10_fraud = new_df['V10'].loc[new_df['Class'] == 1].values
q25, q75 = np.percentile(v10_fraud, 25), np.percentile(v10_fraud, 75)
print(' 25% of v10_fraud : {} | 75% of v10_fraud: {}'.format(q25, q75))
v10_iqr = q75 - q25
print('iqr: {}'.format(v10_iqr))
v10_cutoff = v10_iqr * 1.5
v10_lower, v10_upper = q25 - v10_cutoff, q75 + v10_cutoff
print('Cut Off: {}'.format(round(v10_cutoff,2)))
print('V10 Lower: {}'.format(round(v10_lower,2)))
print('V10 Upper: {}'.format(round(v10_upper,2)))
#condition to remove an outlier
outliers = [x for x in v10_fraud if x < v10_lower or x > v10_upper]
print('Number of outliers in V10 for Fraud Cases: {}'.format(len(outliers)))
print('V10 outliers:', [float(x) for x in outliers])
new_df = new_df.drop(new_df[(new_df['V10'] > v10_upper) | (new_df['V10'] < v10_lower)].index)
#Number of instances after removing the outliers
print('Number of Instances after outliers removal: {}'.format(len(new_df)))




```
